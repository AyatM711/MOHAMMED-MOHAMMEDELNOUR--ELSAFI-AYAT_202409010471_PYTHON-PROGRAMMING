          # Week 3 Documentation: Student Average Calculator

## 1. Identifying the Components

* **Inputs:** 3 quizzes marks
* **Process:** Calculating the average mark of the student and determining if they pass or fail.
* **Output:** Student status (Pass or Fail)

---

## 2. Design the Algorithm

### Pseudocode
```text
START
    SET choice = "y"
    
    WHILE choice converted to lowercase is equal to "y"
        DISPLAY "Enter Quiz 1 mark"
        INPUT quiz_1
        CONVERT quiz_1 to float
        
        DISPLAY "Enter Quiz 2 mark"
        INPUT quiz_2
        CONVERT quiz_2 to float
        
        DISPLAY "Enter Quiz 3 mark"
        INPUT quiz_3
        CONVERT quiz_3 to float
        
        COMPUTE average = (quiz_1 + quiz_2 + quiz_3) / 3
        
        IF average >= 50 THEN
            DISPLAY "Student passed with a mark ", average (formatted to 2 decimal places)
        ELSE
            DISPLAY "student failed with a mark ", average (formatted to 2 decimal places)
        ENDIF
        
        DISPLAY "Countinue? selectY/N:"
        INPUT choice
    ENDWHILE
    
    DISPLAY "program ended"
END
